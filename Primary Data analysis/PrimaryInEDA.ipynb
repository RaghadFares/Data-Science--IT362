```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_primary = pd.read_csv('SurveyData.csv', encoding='ISO-8859-1')
```

## Step 1: Statistical Summaries

```python


# Basic information about the dataset
df_primary.info()

# Descriptive statistics for categorical data
df_primary.describe(include='all')
```

## Step 2: Visualizations

```python

# Plotting the distribution of age groups
plt.figure(figsize=(10, 6))
sns.countplot(y='the age:', data=df_primary, order=df_primary['the age:'].value_counts().index)
plt.title('Age Group Distribution')
plt.xlabel('Count')
plt.ylabel('Age Group')
plt.show()
```

```python


# Plotting the gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender:', data=df_primary)
plt.title('Gender Distribution')
plt.xlabel('Gender:')
plt.ylabel('Count')
plt.show()
```

```python

# Plotting the distribution of hours spent on social media
plt.figure(figsize=(10, 6))
sns.countplot(y='How many hours do you spend on social media platforms daily?', data=df_primary, 
              order=df_primary['How many hours do you spend on social media platforms daily?'].value_counts().index)
plt.title('Hours Spent on Social Media Daily')
plt.xlabel('Count')
plt.ylabel('Hours')
plt.show()
```
